# Analyzing Video with OpenCV and NumPy

## Task 1: Import libraries

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt 

VFILE = "video/hot_rod.mp4"
```

## Task 1: Define generator for reading video

```python
def get_frames(filename):
    video = cv2.VideoCapture(filename)
    while video.isOpened():
        ret, frame = video.read()
        if ret:
            yield frame
        else:
            break
    video.release()
    yield None
```

## Task 2: Retrieve and display video frames

```python
for f in get_frames(VFILE):
    if f is None:
        break
    cv2.imshow('frame', f)
    # Hit Escape to stop
    if cv2.waitKey(10) == 27:
        break
    
cv2.destroyAllWindows()
```

```python
def get_frame(filename, index):
    counter = 0
    video = cv2.VideoCapture(filename)
    while video.isOpened():
        ret, frame = video.read()
        if ret:
            if counter == index:
                return frame
            counter += 1
        else:
            break
    video.release()
    return None
```

```python
frame = get_frame(VFILE, 80)
print('shape ', frame.shape)

# Examine pixel at 0,0
print('pixel at (0,0) ', frame[0,0,:])

plt.imshow(frame)
```

```python
fix_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

# Examine pixel at 0,0
print('pixel at (0,0) ', fix_frame[0,0,:])
plt.imshow(fix_frame)
```

```python
R = fix_frame[:,:,0]
G = fix_frame[:,:,1]
B = fix_frame[:,:,2]
grey = 0.299*R + 0.587*G + 0.114*B
print(grey.shape)
plt.imshow(grey, cmap='gray')
```

```python
# Draw thick circle
frame = get_frame(VFILE, 200)
cv2.circle(frame, 
           center=(200,200), 
           radius=50, 
           color=(0,0,255),
           thickness=10)

fixed_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
plt.imshow(fixed_frame)
```

```python
# Process all video frames
counter = 0
for frame in get_frames(VFILE):
    if frame is None:
        break
    # Add counter
    cv2.putText(frame, 
                text=str(counter), 
                org=(100,100), 
                fontFace=cv2.FONT_HERSHEY_SIMPLEX, 
                fontScale=1, 
                color=(0,255,0),
                thickness=3)
    cv2.imshow('frame', frame)
    
    # Hit Escape to stop
    if cv2.waitKey(10) == 27:
        break
    counter += 1
    
cv2.destroyAllWindows()
```

```python
# FourCC is a 4-byte code used to specify the video codec.
fourcc = cv2.VideoWriter_fourcc('M', 'P', '4', 'V')
video_out = cv2.VideoWriter("new.mp4", fourcc, 20, (640,480))

# Process all video frames
counter = 0
for frame in get_frames(VFILE):
    if frame is None:
        break
    # Add counter
    cv2.putText(frame, 
                text=str(counter), 
                org=(100,100), 
                fontFace=cv2.FONT_HERSHEY_SIMPLEX, 
                fontScale=1, 
                color=(0,255,0),
                thickness=3)
#    cv2.imshow('frame', frame)
    
    # Hit Escape to stop
#    if cv2.waitKey(10) == 27:
#        break
    video_out.write(frame)
    counter += 1
video_out.release()
```

```python
# Get number of frames (requires codec support)
video = cv2.VideoCapture(VFILE)
count = int(video.get(cv2.CAP_PROP_FRAME_COUNT))
video.release()
print('frame count ', count)
```

```python
# Collage
COLLAGE_ROWS = 3
COLLAGE_COLS = 5
FRAME_ROWS = 480
FRAME_COLS = 640

collage_frames = COLLAGE_ROWS * COLLAGE_COLS
skip_frames = count // collage_frames

# Capture the needed frames
frames = []
counter = 0
for f in get_frames(VFILE):
    if counter > 0 and counter % skip_frames == 0:
        frames.append(f)
    counter += 1

# Build collage
TOTAL_ROWS = FRAME_ROWS*COLLAGE_ROWS
TOTAL_COLS = FRAME_COLS*COLLAGE_COLS
out = np.zeros( (TOTAL_ROWS, TOTAL_COLS, 3), dtype=np.uint8 )
idx = 0
for i in range(COLLAGE_ROWS):
    for j in range(COLLAGE_COLS):
        out[FRAME_ROWS*i:FRAME_ROWS*(i+1), FRAME_COLS*j:FRAME_COLS*(j+1), :] = frames[idx]
        idx += 1
plt.imshow(cv2.cvtColor(out, cv2.COLOR_BGR2RGB))
```
